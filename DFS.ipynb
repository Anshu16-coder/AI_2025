```python
def DFS_VISIT(graph,u):

    global time,color,parent,discover_time,finish_time

    color[u]="GREY"
    time+=1
    discover_time[u]=time
    
    for v in graph[u]:
        if color[v]=="WHITE":
            color[v]="GREY"
            parent[v]=u
            DFS_VISIT(graph,v)
    color[u]="BLACK" 
    time+=1
    finish_time[u]=time    
```

```python
def DFS(graph):

    global time,color,parent,discover_time,finish_time

    time=0
    color={u:"WHITE" for u in graph}
    parent={u:None for u in graph}
    discover_time={u:0 for u in graph}
    finish_time={u:0 for u in graph}


    for u in graph:
        if color[u]=="WHITE":
            DFS_VISIT(graph,u)
```

```python
graph = {
    'u': ['v', 'x'],
    'v': ['y'],
    'w': ['y', 'z'],
    'x': ['v'],
    'y': ['x'],
    'z': ['z']  
}

DFS(graph)

print("Discovery Times:", discover_time)
print("Finish Times:", finish_time)
print("Parents:", parent)
     
```
