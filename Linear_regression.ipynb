```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Your original 5-row data ----
data = np.array([
    [2, 22],
    [4, 43],
    [6, 59],
    [5, 50],
    [1, 12]
])

# Separate X and Y
X = np.array([[x, 1] for x in data[:,0]])   # Add bias (x, 1)
Y = data[:,1].reshape(5, 1)

# ---- Linear regression (Normal Equation) ----
XTX = np.dot(X.T, X)
XTX_inv = np.linalg.inv(XTX)
XTY = np.dot(X.T, Y)
w = np.dot(XTX_inv, XTY)

print("Slope (w1):", w[0][0])
print("Intercept (w0):", w[1][0])

# ---- Prediction and Error ----
Y_pred = X.dot(w)
mse = np.mean((Y - Y_pred)**2)
print("Mean Squared Error:", mse)

# ---- Plot best-fit line ----
x_line = np.linspace(0, 7, 100)
y_line = w[0] * x_line + w[1]

plt.scatter(data[:,0], data[:,1], label="Data points")
plt.plot(x_line, y_line, label="Best-fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
