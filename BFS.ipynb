```python
from collections import deque

def BFS(graph,start):

    # Helper dictionaries
    color={u:"WHITE" for u in graph}
    distance={u:float("inf") for u in graph}
    parent={u:None for u in graph}
    
    # Initial setup 
    color[start]="GREY"
    distance[start]=0
    parent[start]=None

    # Queue for BFS
    Q=deque()
    Q.append(start)

    # Main BFS loop 
    while Q:
        u=Q.popleft()
        for v in graph[u]:
            if color[v]=="WHITE":
                color[v]="GREY"
                distance[v]=distance[u]+1
                parent[v]=u
                Q.append(v)
        color[u]="BLACK"        

    return distance,parent
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
distance, parent = BFS(graph, 'A')

print("Distances from source:", distance)
print("Parents in BFS tree:", parent)
```
